```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np
from scipy.optimize import curve_fit

Z=np.array([26,42,28,29,30,40,47,22])
K_alpha=np.array([6.41,17.46,8.01,8.09,8.68,15.81,21.95,4.47])
K_alpha_err=np.array([0.15,0.17,0.16,0.16,0.16,0.17,0.17,0.17])
sqrt_K_alpha=np.sqrt(K_alpha)
sqrt_K_alpha_err=0.5/sqrt_K_alpha*K_alpha_err

n1=1
n2=2

def fit_func(x, sqrt_Er, sig12):
    return sqrt_Er*(x-sig12)*np.sqrt(1/n1**2-1/n2**2)

popt,pcov = curve_fit(fit_func,Z,sqrt_K_alpha,sigma=sqrt_K_alpha_err)

print("sqrt_Er=",popt[0],",Standartfehler=",np.sqrt(pcov[0][0]))
print("sig12=",popt[1],",Standartfehler=",np.sqrt(pcov[1][1]))


plt.errorbar(Z,sqrt_K_alpha,sqrt_K_alpha_err, fmt=".")
plt.plot(Z, fit_func(Z,*popt))
plt.xlabel('Kernladungszahl Z')
plt.ylabel(r'$\sqrt{E_\alpha}/\sqrt{keV}$',fontsize=14)
plt.title('Plot1:'+r'$\sqrt{E_\alpha}$'+' als Funktion von Z')
plt.savefig("K_alpha_vs_Z.pdf", format="pdf")h
```

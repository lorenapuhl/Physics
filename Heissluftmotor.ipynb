```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#Werte
#thermischer Wirkungsgrad
therm=np.array([7.46,8.52,8.92,9.44])
err_therm=np.array([0.08,0.3,0.16,0.21])
#effektiver Wirkungsgrad
eff=np.array([3.46,2.95,2.18,1.24]) #davor 5.08 letzter wert!!!
err_eff=np.array([0.05,0.11,0.06,0.05])
#Frequenz
f=np.array([254.2,276,336.2,341])
err_f=np.array([1.3,9,2.5,1])
```

```python
# plot
plt.errorbar(f, therm, xerr=err_f, yerr=err_therm, fmt='.', ms = 1)
plt.ylabel("thermischer Wirkungsgrad in %")
plt.xlabel("Frequenz in min^-1")
plt.title("thermischer Wirkungsgrad in Abhängigkeit der Frequenz")

#Speichern
fig=plt.gcf()
fig.set_size_inches(11.69, 8.7)
plt.savefig("C:/Users/lorena/Desktop/Heissluftmotor/Diagramm_1.pdf", format="pdf")
```

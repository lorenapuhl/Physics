# Relative motion of two point-like bodies under their mutual gravitational influence. Use a step-by-step Euler integration procedure.

Set $G= 1$, $M1=M2= 1$.

Defining the initial seperation between both bodies as $R_0$

Discretisiing the time evolution using fixed time intervals $ h = \tau_i - \tau_{i-1}$

The separation at a given time interval becomes $\vec{s_i} = \vec{s_{i-1}} + \vec{w_{i-1}}h + O(h^2)$

The characterisitic velocity defined as $\vec{w}=\frac{\vec{v}}{V_0}$ where $V_0 = (\frac{GM}{R_0})^{\frac{1}{2}}$

The velocity at a given time interval follows $\vec{w_i} = \vec{w_{i-1}} - \frac{\vec{s_{i-1}}}{s_{i-1}^3}h + O(h^2)$ 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define time grid
h = 100 # Step size
t = np.arange(0, 10000 + h, h) # time interval

#initial conditions
R0 = 1 
V0 = 1/np.sqrt(R0)
r0 = np.array([1,1])
v0 = np.array([1000,1000])
```

```python
#Euler integration

# Defining arrays
S = np.array([np.array([0,0]) for x in range(len(t))])
W = np.array([np.array([0,0]) for x in range(len(t))])

#initial values

S[0] = r0/R0
W[0] = v0/V0

#Computation
for i in range(1, len(t)):
    W[i] = W[i-1] - (S[i-1]/np.linalg.norm(S[i-1])**3)*h
    S[i] = S[i-1] - W[i-1]*h

print(S[:,0])
print(S[:,1])

#Plot    
plt.figure(figsize = (12, 8))
plt.plot(S[:,0],S[:,1], label = 'orbit')
plt.title('Orbit')
plt.xlabel('sx')
plt.ylabel('sy')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

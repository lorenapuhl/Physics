```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define both sides of the equation
def lhs(r,x,k):
    return r*(1-(x/k))

def rhs(x):
    return (x/(1+x**2))

#Parameters
k = 7.7 #according to exercise
r=0.59

#plot 
x = np.linspace(0.,10.,100)

plt.figure(figsize=(10,8))
plt.plot(x,lhs(r,x,k),label='lhs')
plt.plot(x,rhs(x),label='rhs')
plt.xlabel('x')
plt.ylabel('Rate functions')
plt.title('Rate functions over x')
plt.legend()
plt.show()
```

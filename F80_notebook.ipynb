```python
#plotting the Histogramms
import numpy as np#
import matplotlib.pyplot as plt
import pandas as pd

M = []
Names = ["CS137_1min_42_1.dat",
         "CS137_1min_42_2.dat",
         "caesium137_420V_raw.dat",
         "cobalt60_420V_raw.dat",
         "Sr_background_acryl_test_raw.dat",
         "Str90_meassurement_background_corrected.dat",
         "Str90_meassurement_01_raw.dat",
         "Str90_meassurement_01_acryl_raw.dat",
         "Str90_meassurement_01_0,5mm_alu_raw.dat",
         "Str90_meassurement_01_1mm_alu_raw.dat",
         "Str90_meassurement_01_1,5mm_alu_raw.dat",
         "Str90_meassurement_01_2mm_alu_raw.dat",
         "Str90_meassurement_01_2,5mm_alu_raw.dat",
         "Str90_meassurement_01_3mm_alu_raw.dat",
         "Str90_meassurement_01_3,5mm_alu_raw.dat",
         "Str90_meassurement_01_4,5mm_alu_raw.dat",
         "Str90_meassurement_01_5mm_alu_raw.dat",
         "Str90_meassurement_01_5,5mm_alu_raw.dat",
         "Str90_meassurement_01_6mm_alu_raw.dat",
         "Str90_meassurement_01_6,5mm_alu_raw.dat",
         "Str90_meassurement_01_7mm_alu_raw.dat",
         "Str90_meassurement_01_7,5mm_alu_raw.dat",
         "Str90_meassurement_01_8mm_alu_raw.dat"]

def convert_komma(name):
    f = open(name, 'r')
    if f.mode == 'r':
        data = f.read()
        d = data.split()
        d_2 = [x.replace(',','.') for x in d]
        d_3 = [float(x) for x in d_2]
    return d_3

M = [convert_komma(n) for n in Names]

for x in M:
    print(x)
```

```python
#plotting pulse heights
pulse_height = [959,870,724,557,326,261]
pulse_height_err = [15,13,32,17,25,20]
HV = [451,436,420,405,380,370]

def linear(x,a,b):
    return a*x + b

cf = np.polyfit(HV,pulse_height,1)
x = np.arange(350,470)
fig = plt.figure(figsize = (10.8,8.5))
ax1 = fig.add_subplot(111)
ax1.errorbar(HV,pulse_height,yerr = pulse_height_err, ms = 7,marker = 'x',elinewidth = 0.8, linestyle = 'None',capsize = 3, mew = 0.8, label = 'Pulse Height of the Photo Peak')
ax1.plot(x,linear(x,cf[0],cf[1]), label = 'linear fit to peak positions')
ax1.set_xlabel('HV [V]')
ax1.text('A(U) = ')
ax1.set_ylabel('Pulse height [Bins]')
ax1.set_xlim((350,470))
ax1.set_title("Pulse Height of the Photo Peak")
ax1.grid()
ax1.legend()
plt.show()
```

```python
from IPython.display import IFrame

def gauss(x,a,b,c):
    return c * np.exp(-0.5*((x-b)/a)**2)

from scipy.optimize import curve_fit

popt, pcov = curve_fit(gauss, range(100,500), M[2][100:500],[10,300,250])
popt2, pcov2 = curve_fit(gauss, range(490,580), M[3][490:580],[35,520,85])
popt3, pcov3 = curve_fit(gauss, range(550,650), M[3][550:650],[35,600,50])



print(popt)
print(popt2)
print(popt3)
fig = plt.figure(figsize= (10.8,8.5))
ax1 = fig.add_subplot(111)
ax1.plot(range(0,len(M[3])),M[3], label = '$Co^{60}$ Spectrum',marker = '.')
ax1.plot(range(len(M[2])),M[2],label = '$Cs^{137}$ Spectrum', marker = '.')
ax1.plot(range(200,400),gauss(range(200,400),*popt),label = 'Gauss fit Photopeak Cs $E_{\gamma} = 0.661 MeV$', linewidth = 3)
ax1.plot(range(490,580),gauss(range(490,580),*popt2),label = 'Gauss fit Photopeak Co $E_{\gamma_1} = 1.3325  MeV$', linewidth = 3)
ax1.plot(range(550,700),gauss(range(550,700),*popt3),label = 'Gauss fit Photopeak Co $E_{\gamma_2} = 1.1733 MeV$', linewidth = 3)
ax1.text(230,255,"$\mu$ = 305 $\equiv$ 0.661 MeV")
ax1.text(450,90,"$\mu$ = 531 $\equiv$ 1.3325 MeV")
ax1.text(550,65,"$\mu$ = 602 $\equiv$ 1.1733 MeV")
ax1.set_xlabel('Channel')
ax1.set_ylabel('count')
ax1.set_xlim((0,700))
ax1.set_title('Spectra for Energy calibration')
ax1.legend()
plt.show()
```

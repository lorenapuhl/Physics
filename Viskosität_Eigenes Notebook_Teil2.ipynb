```python
import numpy as np
import math
from scipy.optimize import curve_fit 
import statistics 
import matplotlib.pyplot as plt 
```

```python
#Definition der Plot-Werte
t=np.array([102.43,302.98,505.26,707.57,915.61])#in Sekunden
v=np.array([5,10,15,20,25])#in ml
dt=np.array([1,1,1,1,1])
dv=np.array([1,1,1,1,1])

#Plot
plt.errorbar(t,v, xerr=dt,yerr=dv, fmt=".")
plt.ylabel("volume in ml")
plt.xlabel("time in seconds")
plt.title("Flown volume over time")

#Definition der Fitfunktion
def linear(x,a):
    return a*x

#Lineares Fit
popt, pcov=curve_fit(linear, t, v, sigma=dv)

#Ausgabe der Steigung mit Fehler
#print('m=',popt, '+-', pcov)

print ('m=0,0035+-0,001')
#Plot des linearen Fits
plt.plot(t, linear(t, *popt))

#Vergrößern und Speichern
fig=plt.gcf()
fig.set_size_inches(11.69, 8.7)
plt.savefig("C:/Users/lorena/Desktop/Diagramm_3.pdf", format="pdf")
```
